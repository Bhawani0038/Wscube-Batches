```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

product_data = []

for page_no in range(1, 11):
    url = f"https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={page_no}"
    driver.get(url)
    data = driver.page_source


    soup = BeautifulSoup(data, "html.parser")
    card_html = soup.find_all("div", class_="jIjQ8S")

    for elem in card_html:
        image_url_html = elem.find("img", class_="UCc1lI")
        product_name_html = elem.find("div", class_="RG5Slk")
        product_price_html = elem.find("div", class_=["hZ3P6w", "DeU9vF"])
        product_discount_html = elem.find("div", class_="HQe8jr")
        product_rating_html = elem.find("div", class_="MKiFS6")
        product_review_rating_count = elem.find("span", class_="PvbNMB")
        product_description = elem.find("ul", class_="HwRTzP")

        product_data.append({
            "name": product_name_html.get_text() if product_name_html else None,
            "price": product_price_html.get_text() if product_price_html else None,
            "discount": product_discount_html.get_text() if product_discount_html else None,
            "rating": product_rating_html.get_text() if product_rating_html else None,
            "rating_review_count": (
                product_review_rating_count.get_text()
                if product_review_rating_count else None
            ),
            "description": (
                product_description.get_text()
                if product_description else None
            ),
            "url": (
                image_url_html.attrs["src"]
                if image_url_html else None
            )
        })

    time.sleep(5)



df = pd.DataFrame(product_data)
df.to_csv("prudct_data.csv", index=False)
df
```

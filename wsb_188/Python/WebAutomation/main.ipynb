```python
from textblob import TextBlob
```

```python
import pandas as pd
```

```python
def analyze_sentiment(text):
    blob = TextBlob(text)
    if blob.polarity == 0:
        return "neutral"
    
    elif blob.polarity > 0:
        return "positive"
    
    else:
        return "negative"
```

```python
df = pd.read_clipboard()
```

```python
df['comment_type'] = df['Review Comment'].apply(func= analyze_sentiment)
```

```python
print(df.to_string())
```

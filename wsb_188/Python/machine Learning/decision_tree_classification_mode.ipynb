```python
import pandas as pd

# Create dataset
data = {
    "Customer_ID": ["C001", "C002", "C003", "C004", "C005"],
    "Gender": ["Male", "Female", "Female", "Male", "Male"],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"],
    "Purchased": ["Yes", "No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Copy dataframe to avoid mutation
df_label = df.copy()

label_encoder = LabelEncoder()

# Encode categorical columns
df_label["Gender"] = label_encoder.fit_transform(df_label["Gender"])
df_label["City"] = label_encoder.fit_transform(df_label["City"])
df_label["Purchased"] = label_encoder.fit_transform(df_label["Purchased"])

print("\nLabel Encoded DataFrame:")
print(df_label)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Separate features and target
X = df[["Gender", "City"]]
y = df["Purchased"]



# Define column transformer
ct = ColumnTransformer(
    transformers=[
        ("onehot", OneHotEncoder( sparse_output=False), ["Gender", "City"])
    ],
    # remainder="drop"
)

ct

# Fit and transform
X_encoded = ct.fit_transform(X)
```

```python

# Convert to DataFrame with proper column names
encoded_columns = ct.named_transformers_["onehot"].get_feature_names_out(["Gender", "City"])
X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_columns)

print("\nOne-Hot Encoded Features:")
print(X_encoded_df)

print(X_encoded)
# df
```

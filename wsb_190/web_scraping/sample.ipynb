```python
from bs4 import BeautifulSoup
import requests
import selenium.webdriver as webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
```

```python
import pandas as pd
```

```python
product_list = []
driver = webdriver.Chrome()

for page in range(1, 6):
        driver.get(f"https://www.flipkart.com/search?q=mobilr&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={page}")

        html_code = driver.page_source

        soup = BeautifulSoup(html_code, "html.parser")

        card_html_list = soup.find_all("div", class_="jIjQ8S")

        for card_html in card_html_list:
                image_html = card_html.find("img", class_ = "UCc1lI")
                product_name_html = card_html.find("div", class_ = "RG5Slk")
                start_rating_html = card_html.find("div", class_ = "MKiFS6")
                review_rating_count = card_html.find("span", class_ = "PvbNMB")
                price_html = card_html.find("div", class_ = ["hZ3P6w", "DeU9vF"])
                discount_html = card_html.find("div", class_ = "HQe8jr")
                mobile_discription_html = card_html.find("ul", class_ = "HwRTzP")

                product_list.append({
                        "name"          :  product_name_html.get_text() if product_name_html else None,
                        "url"           :  image_html.attrs['src'] if image_html else None,
                        "rating"        :  start_rating_html.get_text() if start_rating_html else None,
                        "review_count"  :  review_rating_count.get_text() if review_rating_count else None,
                        "price"         :  price_html.get_text() if price_html else None,
                        "discount"      :  discount_html.get_text()  if discount_html else None,
                        "description"   :  mobile_discription_html.get_text() if mobile_discription_html else None
                })


```

```python
df = pd.DataFrame(product_list)
```

```python
df.to_csv('flipkart_mobile_data.csv')
```

```python
from textblob import TextBlob

sentence = "i believe that android is better than ios. I love python"

blob = TextBlob(sentence)

print(blob.sentiment, blob.words, blob.sentences)
```

```python
def sentiment_analyse(sentence):
    blob = TextBlob(sentence)

    if blob.polarity > 0.1:
        return "Positive"
    elif blob.polarity < -0.1:
        return "Negative"
    else:
        return "Neutral"
```

```python
df2 = pd.read_clipboard()
```

```python
df2['sentiment2'] = df2['Review_Comment'].apply(func= sentiment_analyse)
```
